## 1) 프로젝트 목표 및 내용

 이 프로젝트에서는 19/20 시즌에 리그 경기 최소 1000분 이상을 뛰었고 Premier League 상위 7팀에 소속된 축구선수들의 19/20 시즌 성적 데이터를 바탕으로 같은 포지션의 선수들을 비교, 분석하는 프로그램이다. 포지션마다 중요한 스탯들이 각기 다르기 때문에, 각 포지션에 중요한 스탯으로 같은 포지션의 선수들끼리만 비교, 분석할 것이다. 예를 들어, 스트라이커는 득점, 슈팅 등의 스탯을 중점으로 할 것이고, 공격형 미드필더는 찬스메이킹, 패스 등의 스탯을 중점으로 할 것이다. 최종 목표는 크게 3가지가 있다. 첫번째는 팀 스카우터들이 본인의 팀에게 이익을 주는 것이다. 자세히 말하자면, 다른 팀과 경기가 있을 때, 상대 팀의 선수들을 분석해 우리 팀의 승리에 기여할 수 있도록 하거나 혹은 다른 팀에서 선수를 영입하고자 할 때 이 프로젝트가 도움이 될 것이다. 두번째는 중계 방송사에 정보를 주는 것이다. 프리미어리그 경기에서, 양 팀 선수들의 정보를 중계 화면에 올리거나 서로 비교하는 화면을 올릴 때 이 프로젝트가 필요하다. 세번째는 축구팬들에게 정보를 주는 것이다. 축구팬들에게도 다양한 축구 통계 분석 자료들을 볼 권리가 있다. 따라서, 이 프로젝트를 통해 축구팬들에게도 유익한 정보를 줄 수 있다. 이렇게 프로젝트 목표는 총 3가지이며, 해당 목표를 이룰 수 있으려면, 유익한 통계 자료와 시각적인 도구, 적절한 분석들이 잘 구현되어야 할 것이다.

## 2) 주제 선정 이유

이 주제를 선정한 이유는 개인적으로 데이터 분석가가 되고자 하는 것이 목표이고, 평소에 취미로 해외축구를 즐겨봤기 때문에 꼭 한번 만들어보고 싶은 주제였기 때문이다. 해외축구 선수들의 기록들을 보는 것에 대해 흥미가 있었고, Opta, Squawka 등 유명한 해외축구 통계 분석 프로그램을 자주 이용했다. 그러나, 이 주제를 선정한 이유는 이러한 개인적인 부분 뿐만은 아니다. 대부분의 분야에서도 마찬가지겠지만 최근에는 축구계에서도 데이터 분석에 대한 중요성이 대두되고 있다. 요즘 해외에 유명한 축구 프로클럽들은 대부분 팀의 데이터 분석가를 고용할 정도로 데이터 분석이 중요해졌다. 이런 데이터 분석이 팀의 승리에 기여하는 정도가 상당히 커졌을 뿐더러, 전체적인 팀의 이익에 있어서 상당히 중요한 역할을 하고 있다는 것이 요즘 축구계에서 인정되고 있는 부분이다. 전통적 축구 옹호론자들은 이런 통계적인 데이터 분석이 축구 경기에 전혀 쓸모없고 오히려 과격하고 열정적이어야 하는 축구 플레이에 방해가 되고 있다며 거부하지만, 요즈음엔 최신기술 발달에 따라 통계,수학적인 방법을 통한 분석을 통해 팀에 이익을 줄 수 있는 것은 명백한 사실이다. 따라서 주제 선정 이유는 개인적인 흥미 뿐만 아니라 요즘 축구계에서 이러한 데이터 분석이 상당히 중요해졌고, 내가 이런 부분들을 파이썬을 통해 직접 다뤄보고 싶었다. 최근, 국내 스타트업 bepro11이라는 축구 영상 AI 분석 플랫폼을 국내,해외 많은 클럽들이 이용하고 있다. 이렇게, Opta, Squawka, bepro11 등을 알아보고 이용하면서 나 역시 이러한 곳에 참여하고 싶다는 생각을 많이 했었고 그러므로 이런 주제의 프로젝트를 선정하여 프로그램을 작게 개발해보면 이후에 많은 도움이 될 수 있지 않을까하는 생각도 갖게 되었다.

## 5) 분석 결과

### 1. 스트라이커 분석

#### 주요 분석 : 1) 왼발 골 vs 오른발 골, 2) 전체 골 수, 3) 득점 당 시간, 4) 슈팅 정확도, 5) 라이벌 팀끼리의 비교

```python
Striker = pd.read_csv('C:/Users/PC/Desktop/Striker_Info.csv', encoding = 'cp949') #csv파일을 pandas로 읽기
Striker.set_index('Player', inplace = True) #그래프 그리기 위해 Player을 index로 세팅
Striker.sort_values(by = "Player") #csv파일 가나다순으로 전체 출력
```

Striker_Info.csv 파일을 읽어 선수들을 알파벳 순으로 모두 화면에 출력했다.

```python
#헤딩골 비교
Striker['Headed goals'].sort_values().plot(kind='barh', grid=True, figsize=(10,7)) 
#sort_values를 통해 헤딩골을 많이 넣은 순으로 정렬, 막대그래프, gird, 화면크기 설정
plt.title('Headed goals') #그래프 제목 설정
plt.xlabel('Goals') # x축 이름 설정
plt.show() #결과 출력
```

헤딩골 부분에서는 크게 압도하는 선수가 없었다. 칼버트-르윈이 4골로 가장 많은 헤딩골을 기록했고, 스털링, 래시포드, 살라, 라카제트, 손흥민은 1골로 헤딩에 강점이 있는 선수가 아님을 알 수 있었다.  

```python
#왼발 골 비교
Striker['Goals with left foot'].sort_values().plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Goals with left foot')
plt.xlabel('Goals')
plt.show()
```

왼발 골 부분에서는 한 선수가 압도적인 모습을 보여주었다. 살라가 무려 13골을 기록하였던 반면, 에이브라함, 케인, 라카제트는 0~1골로 왼발에서 약점을 보였다.

```python
# 오른발 골 비교
Striker['Goals with right foot'].sort_values().plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Goals with right foot')
plt.xlabel('Goals')
plt.show()
```

오른발 골 부분에서는 아구에로 12골, 래시포드 11골, 오바메양 11골로 많은 득점을 기록했다. 왼발 골을 13골 기록한 살라는 오른발에서는 2골로 가장 낮은 골 수를 보였다. 스트라이커 선수 중 오른발잡이가 많아 왼발 골에 비해 전체적으로 오른발 골에서 많은 득점이 나왔다. 그러나, 단순히 왼발 오른발 골 수로 정확히 어떤 선수가 어떤 발에 강점이 있는지 한눈에 알기는 힘들다.

```python
#왼발 골 vs 오른발 골
plt.figure(figsize = (14,7))
plt.scatter(Striker['Goals with right foot'], Striker['Goals with left foot'], s=50) #scatter(x축, y축, 점 크기)
#scatter를 이용해 그래프에 점으로 표현할 수 있게 함
plt.plot([1,12],[1,12], 'g') #plot(x축, y축, color) (y = x 그래프)
plt.title('Left vs Right') 
plt.xlabel('Goals with right foot')
plt.ylabel('Goals with left foot')

for n in range(15):
    plt.text(Striker['Goals with right foot'][n]*1.01, Striker['Goals with left foot'][n] - 0.5, Striker.index[n], fontsize = 12)
#plt.text(x축, y축, text내용, fontsize)로 그래프에 텍스트를 넣음
plt.grid()
plt.show()
```

x축은 오른발 골, y축은 왼발 골으로 설정해, 오른쪽으로 갈 수록 많은 오른발 골, 위로 갈 수록 많은 왼발 골을 넣었음을 보여주는 결과이다. 녹색 선은 y=x 로, 왼발 골 = 오른발 골을 나타내는 선이다. 따라서 다음 자료로 어떤 선수가 왼발 골과 오른발 골 중 어느 발로 더 많은 골을 기록했는지 한눈에 알 수 있다. 살라는 압도적으로 많은 왼발 골, 히샬리송은 왼발 골 > 오른발 골, 피르미누와 제주스는 왼발 골 = 오른발 골, 나머지 선수들은 모두 왼발 골 < 오른발 골 임을 알 수 있다.

```python
#전체 골 비교
Striker['Goals'] = Striker['Headed goals'] + Striker['Goals with left foot'] + Striker['Goals with right foot']
Striker['Goals'].sort_values().plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Goals')
plt.xlabel('Goals')
plt.show()
```

헤딩 골 + 왼발 골 + 오른발 골로 전체 골 수를 비교하였다. 결과는 오바메양이 17골로 1위, 살라와 아구에로가 16골로 공동 2위를 기록했다.

```python
#득점 당 시간(min) 비교
Striker['Minutes per goal(min)'] = Striker['Minutes played'] / Striker['Goals']
Striker['Minutes per goal(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per goal')
plt.xlabel('min')
plt.show()

#공격 포인트 당 시간(min) 비교
Striker['Minutes per offense points(min)'] = Striker['Minutes played'] / (Striker['Goals'] + Striker['Assists'])
Striker['Minutes per offense points(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per offense points')
plt.xlabel('min')
plt.show()
```

스트라이커 분석 중 가장 의미있는 결과이다. 뛴 시간(분) / 골 을 비교했기 때문에, 선수의 득점이 어느 주기로 나오는지 보여준다. 따라서, 이 결과는 선수들의 득점 생산성을 아주 잘 보여준다. 이 수치는 낮을수록 더 좋은 결과이므로 ascending = False로 두었다. 1위 아구에로는 2위 제주스와 압도적인 기록으로 약 80분에 1골을 득점했다. 1경기 당 1골 이상을 득점한 것이므로 엄청난 기록이라고 볼 수 있다. 반면, 최하위 피르미누는 300분 당 1골, 즉 3경기에 1골을 채 득점하지 못했으므로 득점 부분에서 비생산적인 모습을 보였다. 그러나 앞에서 언급했지만, 피르미누는 리버풀 팀에서 전술적으로 공격형 미드필더에 가까운 역할을 수행한다는 점 또한 고려해야 한다. 뛴 시간/ 공격 포인트도 순위에 차이는 있지만 전체적으로 비슷한 결과이다. 또한, 이 두 결과에서 간과하지 말아야 할 것은 뛴 시간이다. 사실 뛴 시간이 적으면 해당 분석에서 좋은 결과를 가져올 가능성이 높다. 1위 아구에로의 뛴 시간은 1401분으로 그렇게 많은 시간을 소화하지 못했다. 반면, 첫번째 분석에서 최하위인 피르미누와 두번째 분석에서 최하위인 히샬리송의 뛴 시간은 각각 1,2위로 제일 오랜 시간 뛰었음을 기억해야 한다. 따라서, 뛴 시간 / 공격 포인트 부분에서 상위권을 기록한 마네와 살라는 뛴 시간을 고려했을 때 더 인정받아야 할 기록인 것이다.

```python
#슈팅 정확도 비교
Striker['Shooting accuracy(%)'] = Striker['Shots on target'] / Striker['Shots'] * 100
Striker['Shooting accuracy(%)'].sort_values().plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Shooting accuracy')
plt.xlabel('%')
plt.show()
```

슈팅 정확도는 유효 슈팅 / 전체 슈팅으로 어느 확률로 슈팅이 골대를 향했는지를 가리키는 수치이다. 다음 결과에서는 제주스가 50% 이상의 슈팅 정확도를 기록했다. 반면, 피르미누, 아구에로, 히샬리송, 스털링은 40% 이하의 슈팅 정확도를 기록했다.

### 윙어 분석

#### 주요 분석 1) 전체 공격 포인트, 2) 공격 포인트 당 시간, 3) 크로스,패스,슈팅 당 시간, 4) 찬스 메이킹 당 시간

```python
Winger = pd.read_csv('C:/Users/PC/Desktop/Winger_Info.csv', encoding = 'cp949') #csv파일을 pandas로 읽기
Winger.set_index('Player', inplace = True) #그래프 그리기 위해 Player을 index로 세팅
Winger.sort_values(by = "Player") #csv파일 가나다순으로 전체 출력
```

Winger_Info.csv 파일을 읽어 선수들을 알파벳 순으로 나열했다.

```python
#전체 골 비교
Winger['Goals'] = Winger['Headed goals'] + Winger['Goals with left foot'] + Winger['Goals with right foot']
Winger['Goals'].sort_values().plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Goals')
plt.xlabel('Goals')
plt.show()

#전체 공격 포인트 비교
Winger['Offense points'] = Winger['Goals'] + Winger['Assists']
Winger['Offense points'].sort_values().plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Offense points')
plt.xlabel('Offense points')
plt.show()
```

윙어 포지션의 선수들이라 골 수는 크게 중요한 부분은 아니지만, 가장 기본적인 수치이므로 비교하였다. 마레즈가 7골로 가장 많은 득점을 기록했다. 오히려 골보다 골과 어시스트가 합쳐진 공격 포인트가 더 의미있는 수치로 볼 수 있는데, 마레즈가 총 15포인트로 높은 수치를 기록했다.

```python
#득점 당 시간(min) 비교
Winger['Minutes per goal(min)'] = Winger['Minutes played'] / Winger['Goals']
Winger['Minutes per goal(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per goal')
plt.xlabel('min')
plt.show()

#공격 포인트 당 시간(min) 비교
Winger['Minutes per offense points(min)'] = Winger['Minutes played'] / (Winger['Goals'] + Winger['Assists'])
Winger['Minutes per offense points(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per offense points')
plt.xlabel('min')
plt.show()
```

윙어 포지션의 선수들이므로 득점 당 시간 수치보다 공격 포인트 당 시간 수치가 더 중요한 수치라고 볼 수 있다. 1위 마레즈는 무려 약 90분에 공격 포인트 하나를 기록했으며 엄청난 공격 포인트 생산력을 보여주었다. 위 스트라이커 선수들과 비교해도 아구에로, 제주스 다음으로 3위에 해당하는 기록이다. 2000분 가량 상당한 시간을 소화한 5위 윌리안과 7위 제임스는 꽤 선전했다고 볼 수 있겠다. 9위 월콧과 10위 사카는 350분이 넘어가는 수치이므로 공격 포인트를 기록하는 데에 있어 좋은 활약을 하지 못했다고 평가할 수 있다.

```python
#크로스 당 시간(min) 비교
Winger['Minutes per crosses(min)'] = Winger['Minutes played'] / Winger['Crosses']
Winger['Minutes per crosses(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per crosses')
plt.xlabel('min')
plt.show()

#패스 당 시간(min) 비교
Winger['Minutes per passes(min)'] = Winger['Minutes played'] / Winger['Passes']
Winger['Minutes per passes(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per passes')
plt.xlabel('min')
plt.show()

#슛 당 시간(min) 비교
Winger['Minutes per shots(min)'] = Winger['Minutes played'] / Winger['Shots']
Winger['Minutes per shots(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per shots')
plt.xlabel('min')
plt.show()
```

이 분석은 선수에 대한 평가가 아닌 선수에 대한 역할 분석이다. 윙어 포지션에 있는 선수들은 주로 넓은 지역에서 스트라이커에게 크로스를 올리는 역할이 부여되지만, 높고 좁은 지역에서 슈팅을 통한 득점을 하는 역할에 더 비중을 둘 수도 있고, 낮은 지역에서 패스를 통한 플레이메이킹 역할에 더 비중을 둘 수도 있다. 첫번째 결과를 통해 루카스에게 크로스를 올리는 역할이 전혀 부여되지 않았다고 해석할 수 있다. 9위 퓰리시치 또한 그 역할에 큰 비중을 두고 있지 않았음을 알 수 있다. 반면, 윌리안, 페페, 마레즈의 경우 약 20분에 한 번 크로스를 올렸고, 한 경기에 약 4.5번 크로스를 올렸다는 것을 알 수 있다. 두번째 결과를 통해서는 베르나르도가 패스를 통한 플레이메이킹 역할에 비중을 두었음이 보인다. 반면, 제임스와 월콧은 이 역할이 부여되지 않았다고 볼 수 있다. 마지막으로 세번째 결과를 통해서 마레즈와 퓰리시치에겐 슈팅을 자주 때리는 역할을 부여한 반면 사카와 베르나르드가 슈팅을 거의 때리지 않는 역할을 부여받았음을 알 수 있다.

```python
#크로스 vs 슈팅
plt.figure(figsize = (14,7))
Winger['Minutes per shots(min)'] = Winger['Minutes played'] / Winger['Shots']
plt.scatter(Winger['Minutes per crosses(min)'], Winger['Minutes per shots(min)'], s=50) #scatter(x축, y축, 점 크기)
plt.plot([0,150],[0,150], 'g') #plot(x축, y축, color) (y = x 그래프)
plt.title('Crosses vs shots') 
plt.xlabel('Minutes per crosses')
plt.ylabel('Minutes per shots')

Winger['Difference'] = np.abs(Winger['Minutes per crosses(min)'] - Winger['Minutes per shots(min)'])
#일부 선수만 text를 띠우기 위해 Difference 항목을 만들어 numpy 모듈을 사용했다.
Winger_sort = Winger.sort_values(by = 'Difference', ascending = False)
for n in range(5):
    plt.text(Winger_sort['Minutes per crosses(min)'][n]*1.01, Winger_sort['Minutes per shots(min)'][n] - 7, Winger_sort.index[n], fontsize = 12)
#plt.text(x축, y축, text내용, fontsize)
plt.grid()
plt.show()
```

더 쉽게 보기 위해 패스를 제외한 크로스와 슈팅만을 한번에 비교하였다. 사카와 베르나르드, 제임스는 거의 슈팅을 때리지 않았고 루카스는 거의 크로스를 올리지 않았음을 알 수 있다. 

```python
#찬스 메이킹 당 시간(min) 비교
Winger['Minutes per Big chances created(min)'] = Winger['Minutes played'] / Winger['Big chances created']
Winger['Minutes per Big chances created(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per Big chances created')
plt.xlabel('min')
plt.show()
```

이 분석은 시간/찬스메이킹 수치로 찬스를 만들어 낸 주기를 의미한다. 앞선 크로스,패스,슈팅 당 시간 수치는 선수에 대한 평가는 힘들었을지라도, 찬스 메이킹은 윙어 포지션 선수들 누구나에게 꽤 중요한 요소이므로, 충분히 평가가 들어갈 수 있다. 따라서, 앞선 결과에서 비판을 피했던 루카스는 이 분석에서는 비판을 피할 수 없을 것이다. 물론, 루카스는 득점에 집중한 역할이었으므로 찬스메이킹 수치가 낮게 나올 수 있지만, 그렇다고 득점을 많이 기록한 것도 아니기 때문이다. 압도적인 1위 마레즈는 골, 어시스트에 이어 찬스메이킹까지 좋은 결과를 보였다. 앞선 분석에서 중상위권을 유지했던 윌리안, 베르나르도, 페페가 찬스메이킹 부분에서도 그 순위권을 유지했다. 그중에서도 윌리안은 뛴 시간이 많으므로 더 높게 쳐줘야 한다. 흥미로운 것은 어시스트 수가 적었던 월콧과 사카가 찬스메이킹에서 좋은 결과를 받았다. 이는 곧 월콧과 사카가 만들어준 좋은 기회를 동료 플레이어가 득점으로 잘 살리지 못했음을 말해준다.

이 분석은 선수의 슈팅 빈도와 선수의 패스 형태에 대한 분석으로, 공격형 미드필더인 선수가 슈팅을 많이 해 득점을 기록하는데에 더 초점을 두는지, 패스를 통한 기회 창출에 더 초점을 두는지 알게 해준다. 또한, 패스 중 스루 패스를 많이 기록했는지, 롱 패스를 많이 기록했는지 보여준다. 데 브라이너는 슛이든 스루 패스든 롱 패스든 1위를 기록하며 슈팅도 많이 때리고, 어려운 패스도 많이 성공했다. 골과 어시스트를 기록하는 데도 큰 기여를 했는데, 패스로 플레이를 만들고 기회를 주는 역할에도 큰 기여를 했음을 보여준다. 마운트는 패스, 스루패스, 롱패스 수치에서 최하위, 반면 슈팅 수치에서 2위를 기록하며 패스보다 슈팅을 통한 득점을 하는 데에 더 역할 비중을 두고 있음을 알 수 있다. 알리와 시구르드손 또한 이런 역할 비중을 둔 것으로 보인다. 한편 외질은 슈팅을 거의 때리지 않은 반면, 패스 수치에서 3위를 기록하며 패스를 통한 기회 창출에 역할을 두었다. 다비드 실바 또한 슈팅보다 패스에 비중을 둔 것으로 보인다. 외질은 막상 어려운 패스들은 거의 기록하지 못한 것을 보아 대부분 쉬운 패스를 시도했음을 알 수 있다.

```python
#찬스 메이킹 당 시간(min) 비교
AttackingMid['Minutes per Big chances created(min)'] = AttackingMid['Minutes played'] / AttackingMid['Big chances created']
AttackingMid['Minutes per Big chances created(min)'].sort_values(ascending = False).plot(kind='barh', grid=True, figsize=(10,7))
plt.title('Minutes per Big chances created')
plt.xlabel('min')
plt.show()
```

찬스 메이킹 능력은 공격형 미드필더에게 가장 중요한 수치라고 볼 수 있다. 찬스 메이킹 당 시간에서도 데 브라이너가 압도적인 수치를 보였다. 데 브라이너는 공격형 미드필더가 갖추어야 할 모든 수치에 1위를 기록하며 19/20 시즌 엄청난 활약을 했다. 다비드 실바도 다른 스탯에서와 같이 좋은 수치를 기록했다. 이전에선 다소 하위권이었던 안드레아스 페레이라는 찬스 메이킹에서는 좋은 활약을 보였다. 그러나, 외질과 체임벌린은 찬스 메이킹을 거의 보여주지 못했으며, 해당 시즌 부진했음을 바로 알 수 있다.

## 6) 결론

스트라이커, 윙어, 공격형 미드필더 세 포지션 선수들을 19/20 시즌 성적을 바탕으로 각 포지션마다 중요한 스탯으로 비교, 분석하였다. 각 영역에 따라 막대그래프를 통해 한 눈에 어떤 선수가 더 좋은 기록을 보여주었는지 알 수 있도록 했다. 필요에 따라 두 영역을 비교할 때에는 점 그래프를 사용했다. 이 분석으로 19/20 시즌 한정해서 어떤 선수가 어느 영역에서 어떠한 모습을 보여주었는지는 확실히 보여주었다고 생각한다. 그러나 아쉬운 것은, 오직 수치만으로 선수에 대한 정확한 평가가 이루어지기에는 쉽지 않았다는 것이다. 실제 축구에서의 데이터 분석은 단편적인 수치보다, 경기 내내의 선수 플레이를 영상 분석을 통해 이루어지며, 그렇게 이루어져야 더 정확한 분석이 가능하다. 일단 여기에선 중요한 수치만을 가지고 분석을 하는 것이므로, 선수에 대해 정확히 평가를 내리기는 사실 쉽진 않다. 경기 내에서의 전체적인 플레이와 선수들에게 부여된 역할들이 골고루 고려가 되어야 한다. 그래도 주어진 수치로 분석할 수 있는 부분은 최대한 시도를 했고, 그 안에서 의미있는 분석들도 충분히 보여졌을 거라고 생각한다.
